```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
```

```python
df = pd.read_csv('Measurement_summary.csv')
df.head()
```

```python
list_add = list(df['Address'])
District = [i.split(', ')[2] for i in list_add]
df['District'] = District
```

```python
list_district = list(set(District))
```

```python
list_district
```

```python
list_YM = [i.split(" ")[0][:-3] for i in  list(df['Measurement date'])]
list_Year = [i.split(" ")[0][0:4] for i in  list(df['Measurement date'])]
list_Month = [i.split(" ")[0][5:7] for i in  list(df['Measurement date'])]

df['YM'] = list_YM
df['Year'] = list_Year
df['Month'] = list_Month
```

```python
df.head()
```

```python
df.info()
```

#### Kleine Exkursion, wie GROUPBY funktioniert

```python
data = pd.DataFrame({'x1':[3, 5, 1, 2, 5, 7, 2, 8],     # Create pandas DataFrame
                     'x2':range(1, 9),
                     'group1':['A', 'B', 'A', 'C', 'A', 'C', 'B', 'A'],
                     'group2':['a', 'a', 'a', 'a', 'b', 'b', 'b', 'b']})
print(data)  
```

```python
data_group = data.groupby(['group1', 'group2']).mean()  # Get mean by group
print(data_group)                                       # Print aggregated data
```

```python
data_group_new = data_group.reset_index()               # Reset index of grouped data
print(data_group_new) 
```

```python
df_monthly = df.groupby(['Station code', 'District', 'YM', 'Year', 'Month'])[['SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5']].mean()
df_monthly
```

```python
df_monthly = df.groupby(['Station code', 'District', 'YM', 'Year', 'Month'])[['SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5']].mean().reset_index()
df_monthly
```

```python
# Set the renderer to 'notebook'
pio.renderers.default = 'iframe'
```

```python
# Farbpaletten in Plotly
fig = px.colors.qualitative.swatches()
fig.show()
```

```python
pal = px.colors.qualitative.Dark24
print(type(pal))
```

#### Graph 1: line chart

```python
fig = go.Figure()
for d,p in zip(list_district, pal):
    fig.add_trace(go.Scatter(x = df_monthly[df_monthly['District']==d]['YM'],
                             y = df_monthly[df_monthly['District']==d]['PM2.5'],
                             name = d,
                             line_color = p, 
                             fill=None))  #tozeroy 
            
fig.show()
```

```python
fig = go.Figure()
for d,p in zip(list_district, pal):
    fig.add_trace(go.Scatter(x = df_monthly[df_monthly['District']==d]['YM'],
                             y = df_monthly[df_monthly['District']==d]['PM2.5'],
                             name = d,
                             line_color = p, 
                             fill='tozeroy'))  #tozeroy 
            
fig.show()
```

#### Graph 3: Facet Grid

```python
fig = px.line(df_monthly, y="SO2", x="YM", facet_col="District", facet_col_wrap=4)
fig.show()
```

```python
fig = px.bar(df_monthly, x="District", y="PM2.5", facet_col="Year", facet_row="Month", facet_col_wrap=4, height=2000, width=1000)
fig.show()
```

#### Graph 4: Heat map

```python
df_pivot = pd.pivot_table(df_monthly,
                          values='PM2.5',
                          index='District',
                          columns='YM')
df_pivot
```

```python
heatmap_trace=go.Heatmap(z=df_pivot)
fig = go.Figure(data=[heatmap_trace])
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Create a DataFrame with the matrix data
data = {
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12],
    'D': [13, 14, 15, 16]
}
df = pd.DataFrame(data)

# Create a trace for the heatmap
heatmap_trace = go.Heatmap(z=df)

# Create the layout
layout = go.Layout(title='Matrix Heatmap')

# Create the figure
fig = go.Figure(data=[heatmap_trace], layout=layout)

# Display the figure
fig.show()
```
